```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Why not have fun?
plt.xkcd()
```

Generate random number with `np.random.normal()`, drawing from normal distribution

```python
# Create random numbers, make 250 datapoints for each x and y
x_ = np.random.normal(size = 250)
y_ = np.random.normal(size = 250)

# Generate categorical value for colormapping in matplotlib, assign int, max = 25 different ints, same size
c_ = np.random.randint(low = 1, high = 25, size = 250)

# Turn these arrays into Pandas DataFrame, using dict notation
df = pd.DataFrame({
    'x_': x_,
    'y_': y_,
    'c_': c_
})
```

I just learned that I could actually call the `.plot()` method from `Axes` object. So there we go, I generated subplots with `plt.subplots()` call, then calling individual Axes to generate the chart.

Lastly, for an authentic scientific touch, I called the `plt.xkcd()` to create a deeper meaning to my scientific illustration. Ahaks.

```python
# Instantiate the Figures and Axes objects
fig, axes = plt.subplots(ncols = 2, figsize = (12, 4), sharey = True)

# Generate first plot at axes[0], no color designation
axes[0].plot(df['x_'], df['y_'], marker = 'o', 
             linestyle = '', markerfacecolor = 'gray')

# Perfrom pd.groupby() to prepare data as groups based on c_ values
groups = df.groupby(c_)

# Looper function to generate the 2nd plot at axes[1]
for name, group in groups:
    axes[1].plot(group['x_'], group['y_'],marker = 'o', 
             linestyle = '', label = name)

# Customizations
axes[0].set_title('Argh we need labels!')
axes[1].set_title('Argh this is not helpful!')

# Save the plot
#fig.savefig('scatterplot_random_label.png', format = 'png', dpi = 300, bbox_inches = 'tight')
```

```python
# Extras: generating chart with colormap for funsies

# Save the generated plt.scatter() into points variable
points = plt.scatter(df['x_'], df['y_'], c = df['c_'], cmap = 'PuRd')

# Call the colormap
cmap = plt.colorbar(points)
```
